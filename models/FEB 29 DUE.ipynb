```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score, mean_squared_error
from joblib import dump, load

# Assuming df is your original DataFrame
# If 'HARVEST SEASON' is not already a string, convert it
df['HARVEST SEASON'] = df['HARVEST SEASON'].astype(str)

# Select relevant features
features = ['day', 'month', 'year']
targets = ['HARVEST SEASON', 'PRICE (PHP per kg)']

# Handle missing values
df = df.dropna(subset=targets + features)
df = df.fillna(method='ffill')

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    df[features],
    df[targets],
    test_size=0.2,
    random_state=42
)

# Define the multi-output model
multi_output_model = RandomForestRegressor(random_state=42)

# Define the preprocessor
numeric_features = ['day', 'month', 'year']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), [])
    ])

# Create the multi-output pipeline
multi_output_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', multi_output_model)
])

# Train the model
multi_output_pipeline.fit(X_train, y_train)

# Make predictions on the test set
predictions = multi_output_pipeline.predict(X_test)

# Unpack the predictions
classification_predictions, regression_predictions = predictions[:, 0], predictions[:, 1]

# Evaluate the classification model
accuracy = accuracy_score(y_test['HARVEST SEASON'], classification_predictions.round())
print(f'Classification Accuracy: {accuracy}')

# Evaluate the regression model
mse = mean_squared_error(y_test['PRICE (PHP per kg)'], regression_predictions)
print(f'Regression Mean Squared Error: {mse}')

# Save the model
dump(multi_output_pipeline, 'multi_output_model.joblib')

# Now you can use this trained model for future predictions as well.
```
